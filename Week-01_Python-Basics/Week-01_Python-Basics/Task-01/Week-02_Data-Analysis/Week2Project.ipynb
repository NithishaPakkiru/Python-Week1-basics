**Student Grade Calculator**

```python
score = int(input("Please enter your score in exam: "))

# Check for invalid input first
if score < 0 or score > 100:
    print("Invalid score! Please enter between 0 and 100.")

# Now check valid score ranges
elif score >= 90:
    print("Grade: A")
    print("Excellent work!")

elif score >= 80:
    print("Grade: B")
    print("Great job! Keep improving.")

elif score >= 70:
    print("Grade: C")
    print("Good effort! You can do even better.")

elif score >= 60:
    print("Grade: D")
    print("Passed, but try to improve.")

elif score >= 40:
    print("Grade: E")
    print("Don't give up, study a bit more.")

else:
    print("You failed. Don't get discouraged â€” try harder next time!")
```

```python
## following code gives different messages based on the current hour of the day

from datetime import datetime
time = datetime.now()
hour = time.hour
print(hour)
```

```python
if hour >= 6 and hour <= 11:
  print("It's Morning")
elif hour >= 12 and hour <= 16:
  print("It's Afternoon")
elif hour >= 17 and hour <= 20:
  print("It's Evening")
elif hour >= 21 or hour <= 5:
  print("It's night")
```

```python
## Number guessing game using if-else statements

actual_number = 5
Guess = int(input("Enter the number: "))
```

```python
while Guess != actual_number:
  if Guess > actual_number:
    print("the number is too high")
  elif Guess < actual_number:
    print("the number is too low")
  Guess = int(input("Try Again: "))

print("Your guess is correct!")
```

```python
## finding the area of a rectangle

def area_of_rectangle(length, breadth):
  area = length * breadth
  return area

print(area_of_rectangle(2,4))
```

```python
## building a calculator to perform addition, subtraction, multiplication and division operations.

x = int(input("give the first number: ") )
y = int(input("give the second number: "))
operation = input("Which operation you want to perform: ")
if operation == '+':
  addition = x+y
  print("Addition of given numbers is: ", addition)
elif operation == '-':
  substraction = x-y
  print("Substraction of given numbers is: ", substraction)
elif operation == '*':
  multiplication = x*y
  print("Multiplication of given numbers is: ", multiplication)
elif operation == '/':
  if y == 0:
    print("Error: Cannot divide by zero")
  else:
    division = x / y
    print("Division of given numbers is:", division)


```
